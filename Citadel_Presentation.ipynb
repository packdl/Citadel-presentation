7. Batteries included.

```python
""" 
    Let's see how many members are in our CSV file 
    and the most used email domain and least used domain.
"""
import csv
from collections import Counter

c = Counter()
with open('members.csv','r') as FILE:
    reader = csv.DictReader(FILE)
    members = list(reader)
    print(f"There are {len(members)} total members.")
    for member in members:
        domain = member["Email Address"].split("@")[-1]
        c[domain]+=1
m_c_domains = c.most_common()
mcd, lcd = m_c_domains[0], m_c_domains[-1]
print(f"The most common domain is {mcd[0]} with {mcd[1]} member")
print(f"The least common domain is {lcd[0]} with {lcd[1]} member")
```

# How are people using it?

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import csv
with open('wordcloud.csv','r') as WCFILE:
    word_freq = list(csv.reader(WCFILE))
    word_freq_dict = {row[0]:int(row[1]) for row in word_freq[1:]}        
cloud = WordCloud(max_font_size=40, height=150,width=300,background_color='white').generate_from_frequencies(word_freq_dict)
```

```python
plt.imshow(cloud, interpolation='bilinear');plt.axis("off");plt.show()
```
